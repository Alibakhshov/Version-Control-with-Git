<h4><b>Git Log</b></h4>
<p>The <b>git log</b> command is used to display all of the commits of a repository.</p>
<p>By default, this command displays:</p>
<ul>
    <li>the SHA</li>
    <li>the author</li>
    <li>the date</li>
    <li>and the message</li>
</ul>



<p>We've been looking closely at all the detailed information that <b>git log</b> displays. But now, take a step back and look at all of the information as a whole.

Let's think about some of these questions:</p>

<ul>
    <li><b>the SHA</b>  - git log will display the complete SHA for every single commit. Each SHA is unique, so we don't really need to see the entire SHA. We could get by perfectly fine with knowing just the first 6-8 characters. Wouldn't it be great if we could save some space and show just the first 5 or so characters of the SHA?</li>
    <li><b>the author</b> - the git log output displays the commit author for every single commit! It could be different for other repositories that have multiple people collaborating together, but for this one, there's only one person making all of the commits, so the commit author will be identical for all of them. Do we need to see the author for each one? What if we wanted to hide that information?</li>
    <li><b>the date</b> - By default, git log will display the date for each commit. But do we really care about the commit's date? Knowing the date might be important occasionally, but typically knowing the date isn't vitally important and can be ignored in a lot of cases. Is there a way we could hide that to save space?</li>
    <li><b>the commit message</b> - this is one of the most important parts of a commit message...we usually always want to see this</li>
</ul>

```python
print("Two Terminal applications side-by-side. The left one shows the result of the git log command with all of the information while the right one shows the result of the git log --stat command which lists the files that were changed as well as the number of added/removed lines")
from IPython.display import Image
Image(filename='Images/git log vs git log --stat.png') 
```

<ul>

<h4><b>Annotated <i>git log -p</i> Output</b></h4>
<p>Using the image above, let's do a quick recap of the <b>git log -p</b> output:</p>
<ul>
    <li>🔵 - the file that is being displayed</li>
    <li>🔶 - the hash of the first version of the file and the hash of the second version of the file</li>
    <ul>
        <li>not usually important, so it's safe to ignore</li>
    </ul>
    <li>❤️ - the old version and current version of the file</li>
    <li>🔍 - the lines where the file is added and how many lines there are</li>
    <ul>
        <li>-15,83 indicates that the old version (represented by the -) started at line 15 and that the file had 83 lines</li>
        <li>+15,85 indicates that the current version (represented by the +) starts at line 15 and that there are now 85 lines...these 85 lines are shown in the patch below </li>
    </ul>
    <li>✏️ - the actual changes made in the commit</li>
    <ul>
        <li>lines that are red and start with a minus (-) were in the original version of the file but have been removed by the commit</li>
        <li>lines that are green and start with a plus (+) are new lines that have been added in the commit
</li>
    </il>
</ul>

<p><b><i>git log -p</i></b> command adds the following to the default output:</p>

<ul>
    <li>displays the files that have been modified</li>
    <li>displays the location of the lines that have been added/removed</li>
    <li>displays the actual changes that have been made
</li>
</ul>

<i> There are two ways to display a specific commit's details:</i>
<ul>
    <li>providing the SHA of the commit you want to see to <b>git log</b></li>
    <li>use a new command <b>git show</b></li>
</ul>

<p>By supplying a SHA, the <b>git log -p</b> command will start at that commit! No need to scroll through everything! Keep in mind that it will also show all of the commits that were made prior to the supplied SHA. Example: <b>git log -p fdf5493</b> </p>

<h4><b>New Command: <i>git show</i></b></h4>
<p>The other command that shows a specific commit is <b>git show</b>:</p>
<h4><b>What does <i>git show</i> do?</b></h4>
<p>The <b>git show</b> command will show only one commit. So don't get alarmed when you can't find any other commits - it only shows one. The output of the <b>git show</b> command is exactly the same as the <b>git log -p</b> command. So by default, <b>git show</b> displays:</p>

<ul>
    <li>the commit</li>
    <li>the author</li>
    <li>the date</li>
    <li>the commit message</li>
    <li>the patch information</li>
</ul>

<p>However, <b>git show</b> can be combined with most of the other flags we've looked at:</p>

<ul>
    <li><b>--stat</b> - to show the how many files were changed and the number of lines that were added/removed</li>
    <li><b>-p</b> or <b>--patch</b> - this the default, but if <b>--stat</b> is used, the patch won't display, so pass <b>-p</b> to add it again</li>
    <li><b>-w</b> - to ignore changes to whitespace</li>
</ul>
