<h4><b>Navigating The Log</b></h4>
<p>If you're not used to a pager on the command line, navigating in Less can be a bit odd. Here are some helpful keys:</p>
<ul>
    <li>to scroll <b>down</b>, press</li>
    <ul>
        <li><b>j</b> or <b>↓</b> to move down one line at a time</li>
        <li><b>d</b> to move by half the page screen</li>
        <li><b>f</b> to move by a whole page screen</li>
    </ul>
    <li> to scroll <b>up</b>, press</li>
    <ul>
        <li><b>k</b> or ↑ to move _up_ one line at a time </li>
        <li><b>u</b>  to move by half the page screen</li>
        <li><b>b</b>to move by a whole page screen</li>
    </ul>
    <li>press <b>q</b> to quit out of the log (returns to the regular command prompt) </li>
</ul>
